```python
import re
```

```python
def clean_citation_text(text):
    # 1. Remove <cite>
    text = text.replace("<cite>", " ")
    # 2. Lowercase
    text = text.lower()
    # 3. Remove weird symbols but keep letters, numbers, punctuation
    text = re.sub(r"[^a-z0-9.,;:()\- ]", " ", text)
    # 4. Fix broken multiple spaces
    text = re.sub(r"\s+", " ", text)
    # 5. Strip leading/trailing spaces
    text = text.strip()
    return text
```

```python
df_train_30["clean__citation_context"] = df_train_30["CITATIONS_CONTEXTS"].apply(clean_citation_text)
```
