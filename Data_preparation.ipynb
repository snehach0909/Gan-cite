```python
import os
import pandas as pd
from sklearn.model_selection import StratifiedKFold, train_test_split

# ── Config ──────────────────────────────────────────────
TRAIN_PATH   = "processed_data/df_train_30_clean.csv"
VAL_PATH     = "processed_data/df_val_30_clean.csv"
TEST_PATH    = "processed_data/df_test_30_clean.csv"

OUTPUT_DIR   = "./dataset/10-fold_labeled_increasing_relabel_ver3"
LABEL_COL    = "UTILIZE_LABELS"
RANDOM_STATE = 42
N_FOLDS      = 10
# ────────────────────────────────────────────────────────

os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load and combine into one pool
train_df = pd.read_csv(TRAIN_PATH)
val_df   = pd.read_csv(VAL_PATH)
test_df  = pd.read_csv(TEST_PATH)

full_df = pd.concat([train_df, val_df, test_df], axis=0).reset_index(drop=True)
print(f"Total pool: {len(full_df)} rows")
print(full_df[LABEL_COL].value_counts())

# Create 10 stratified folds
skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)
X   = full_df.index.values
y   = full_df[LABEL_COL].values

for fold_num, (train_val_idx, test_idx) in enumerate(skf.split(X, y), start=1):

    # Split train_val → train 80% / val 20%
    train_idx, val_idx = train_test_split(
        train_val_idx,
        test_size=0.2,
        random_state=RANDOM_STATE,
        stratify=y[train_val_idx]
    )

    full_df.iloc[train_idx].reset_index(drop=True).to_csv(
        f"{OUTPUT_DIR}/algocite_utilize_dataset_train_fold_{fold_num}.csv", index=False)
    full_df.iloc[val_idx].reset_index(drop=True).to_csv(
        f"{OUTPUT_DIR}/algocite_utilize_dataset_val_fold_{fold_num}.csv", index=False)
    full_df.iloc[test_idx].reset_index(drop=True).to_csv(
        f"{OUTPUT_DIR}/algocite_utilize_dataset_test_fold_{fold_num}.csv", index=False)

    print(f"Fold {fold_num}: train={len(train_idx)}, val={len(val_idx)}, test={len(test_idx)}")

print("\nDone! Files saved to:", OUTPUT_DIR)
```
